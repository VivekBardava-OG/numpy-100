#### 22. Normalize a 5x5 random matrix (★☆☆)

````python
ran_matrix = np.random.randint(1,50,(5,5))
print(ran_matrix)
mean = ran_matrix.mean()
std = ran_matrix.std()
normalized = (ran_matrix - mean) / std
print('\nThe normalized matrix : \n')
print(normalized)
````

#### 27. Consider an integer vector Z, which of these expressions are legal? (★☆☆)
```python
Z**Z
2 << Z >> 2
Z <- Z
1j*Z
Z/1/1
Z<Z>Z
```

````python
import numpy as np
Z = np.array([1,2,3,4,5,6])
print(Z*Z)
print(2 << Z >> 2)
print(Z <- Z)
print(1j*Z)
print(Z/1/1)
# print(Z<Z>Z) not legal
````
